```python
import json
import pyzstd
import re
from collections import Counter
```

```python
# This is hack that tells the system that these files were compressed with
# a very long compression window.
PARAMS = {pyzstd.DParameter.windowLogMax: 31}
# decompress file using pyzstd

ageG_list = [] #empty list to add each ageGender for later counting
age_list = []
with pyzstd.open('RS_2023_12_relationships.zst', "rt", level_or_option=PARAMS) as rships:
      for line in rships: #json lines counted separately
          post = json.loads(line)  # A dictionary with a "title" field.

          #deserialize dictionary first, apply lower case than convert back to dictionary
          dict_in_str = json.dumps(post)
          post_lower = json.loads(dict_in_str.lower()) #normalize data
          # print(post_lower)
          title = post_lower["title"]
          # print(title) #check data


          # genderAge = r'((f|m|nb)(\d{1,2}))'
          # ageGender = r'((\d{1,2})(f|m|nb))'
          # re.finditer(genderAge, title) #find pattern via variable?
          #https://stackoverflow.com/questions/6930982/how-to-use-a-variable-inside-a-regular-expression

#           # ((\d{1,2})\s?(f|m|nb))|((f|m|nb)(\s?|-?)(\d{1,2})) # for "34 f" space cases or 34-f

# could have used "re.IGNORECASE flag" to normalize gender??
## matches = re.findall(pattern, title, re.IGNORECASE)

          ageG = re.finditer(r'((\d{1,2})(f|m|nb))|((f|m|nb)(\d{1,2}))', title)
          if ageG: #if finds match
            for match in ageG:
                phrase = match.group(0)
                ageG_list.append(phrase)
          # finditer = callable iterator object - can't just work w/ the matches found
          # finditer gives a lot of info from matched object
            #group element in this object - manipulate this piece for using data
            ##finditer returns a match object, not just a substring.  need to use group() method on the match object.

          #matches any digit, between 1-2 times (greedy match)
          age = re.finditer(r'(\d{1,2})', title)
          if age: #if finds match
            for match in age:
                phrase = match.group(0)
                age_list.append(phrase)



          # for element, count in Counter(age).most_common(1):
          #     elementInt = int(element)
          #     print((elementInt, count))
                # print(("the most common age is", element, "count:", count))

          #convert the age string into an integer.
          # ageInt= int(age)
          # ageCounter=Counter(ageInt)
          # ageCounter.most_common(1)
```

```python
#most_common method to get the most common elements and their counts
#store the ages and age/gender tuples, Use the built-in collections.Counter
print("the most common age/gender and count:", Counter(ageG_list).most_common(1))
```

```python
# another method of looking for most common age
ages = ageG_list.copy()
cleanAge = []
for entry in ages:
  if 'm' in entry:
    entry = entry.replace('m', '')
  if 'f' in entry:
    entry = entry.replace('f', '') #needs replacement in second argument
  if 'nb' in entry:
    entry = entry.replace('nb', '')
  cleanAge.append(entry)


print("the most common age and count:", Counter(cleanAge).most_common(1))
```

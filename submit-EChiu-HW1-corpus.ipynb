```python
#use Python's built-in csv module to load the provided frequency data.
import csv
import pandas as pd
import math
from collections import Counter
```

```python
file_path = 'kjv.tsv'
df = pd.read_csv(file_path, sep='\t', names=['word', 'count'])

kjv_dict = {}
for ix, row in df.iterrows():
    word = row['word']
    count = row['count']

    kjv_dict[word] = count

kjv_dict
```

```python
c_kjv = Counter(kjv_dict)

c_kjv.total()
#N number of tokens (sum of values)
```

```python
kjv_count = c_kjv.total()
kjv_count
```

```python
file_path = 'horoscopes-1.tsv'
df = pd.read_csv(file_path, sep='\t', names=['word', 'count'])

horo_dict = {}
for ix, row in df.iterrows():
    word = row['word']
    count = row['count']

    horo_dict[word] = count

horo_dict
```

```python
c_horo = Counter(horo_dict)

c_horo.total()
#N number of tokens (sum of values)
```

```python
horo_count = c_horo.total()
horo_count
type(horo_count)
```

```python
## log odds
#lambda keyword is used to create anonymous function
common_keys = filter(lambda x: x in horo_dict, kjv_dict)
# built-in function that builds a new sorted list from an iterable
for key in sorted(common_keys):
  #	log odds = log c - log(N âˆ’ c) (subtraction is division in log)
    print("horoscope log odds for", key, (math.log(horo_dict[key])) - math.log(horo_count - (horo_dict[key])))
    # horo_c1 = horo_dict[key] #count of each horo key
    # kjv_c2 = kjv_dict[key] #count of each kjv key
    # print("the common word is", key, "horoscope value is", horo_c1, "kjv value is", kjv_c2)
```

```python
#approach
#new list, append pair of key & log odds ratio (all log odds ratio & key)
#or dictionary, key & log odds ratio (if close to 0, then occurs equally in both texts)

  #read 2 dictionaries in, find words share, for every word they share, compute log odds ratio (log of two odds)
  #N will be constant, c1 will change by key; c1 and c2 as variables
  #LOR variable, append tuples to list (list outside of loop)- add new pair to list after going through
  ## sort after list/values (first and last, print out 10)
```

```python
# create log odds ratio empty list
lorlist = []
#lambda keyword is used to create anonymous functions
common_keys = filter(lambda x: x in horo_dict, kjv_dict)
# built-in function that builds a new sorted list from an iterable
for key in sorted(common_keys):
    horo_c1 = horo_dict[key] #count of each horo key
    kjv_c2 = kjv_dict[key] #count of each kjv key
    #log odds ratio = log c1 + log(N2 âˆ’ c2) âˆ’ log c2 âˆ’ log(N1 âˆ’ c1)
    LOR = (math.log(horo_c1)) + (math.log(kjv_count - kjv_c2)) - (math.log(kjv_c2)) - (math.log(horo_count - horo_c1))
    # round to 4 decimal places
    LOR = round(LOR, 4)
    # append tuple to empty list (initiated outside of for loop)
    lorlist.append(tuple([key, LOR]))
    print (lorlist) # print statement inside loop for why it's showing multiple times-list growing (keep inside when debugging)
    # take out of loop for printing once at the end
```

```python
#Use the list .sort method (or the sorted function if necessary)
##and its key= keyword argument to sort the words by log-odds ratio.
#sort list of tuples, x[1] = sorting on second position
sorted_list = sorted(lorlist, key=lambda x: x[1])
print(sorted_list)
```

```python
#Then, use subscripting (i.e., the square bracket syntax) to get the highest and lowest ten values.
# Initializing the list
n = 10
lo_res = []
# print original list
print('The log odds ratio list is:', sorted_list)

for i in range(0, n):
    lo_res.append(sorted_list[i])

# printing the n smallest ratios in the list
#(those that are most overrepresented in the Bible)
print('The ', n, ' lowest log-odds ratio are:', lo_res)
```

```python
sortreverse_list = sorted(lorlist, key=lambda x: x[1], reverse=True)
print(sortreverse_list)
```

```python
# Initializing the list
n = 10
hi_res = []
for k in range(0, n):
    hi_res.append(sortreverse_list[k])

# printing the n largest elements in the list
#i.e., those that are the most overrepresented in the horoscopes)
print('The ', n, ' highest log-odds ratios are:', hi_res)
```
